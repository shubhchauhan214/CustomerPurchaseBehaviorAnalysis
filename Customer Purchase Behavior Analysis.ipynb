```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

# Random seed for reproducibility
num_customers = 1000

# Customer IDs
customer_ids = np.arange(1, num_customers + 1)

# Ages between 18 and 65
ages = np.random.randint(18, 66, size=num_customers)

# Total purchases made by each customer
purchases = np.random.poisson(lam=5, size = num_customers)

#Total Amount spent (500 to 5000 per purchase)
amount_spent = purchases * np.random.randint(500, 5000, size = num_customers)

# Repeat customers flag (1 = yes , 0 = no)
repeat_customers = np.random.choice([0,1], size=num_customers, p=[0.3,0.7])

# Months in which purchases were made
purchase_months = np.random.choice(np.arange(1,13), size=num_customers)
```

```python
# Total Revenue
total_revenue = np.sum(amount_spent)
print(f" Total Revenue: ${total_revenue}")

# Average Spending per Customer
avg_spending = np.mean(amount_spent)
print(f" Average Spending per Customer: ${avg_spending:.2f}")

# Total Number of Repeat Customers
total_repeat = np.sum(repeat_customers)
print(f"Total Repeat Customers: {total_repeat}")

# Age Group Distribution 
age_bins = [18, 25, 35, 45, 55, 65]
age_groups = np.digitize(ages, bins = age_bins)
unique, counts = np.unique(age_groups, return_counts=True)

# Display age group stats
print("\n Age Group Distribution:")
for i, count in zip(unique, counts):
    if i == 0:
        label = f"< {age_bins[0]}"
    elif i == len(age_bins):
        label = f"{age_bins[-1]}+"
    else:
        label = f"{age_bins[i-1]} - {age_bins[i]}"
    print(f"{label}: {count} customers")
```

```python
age_labels = ['18-25', '25-35', '35-45', '45-55', '55-65', '65+']
age_counts = counts

plt.figure(figsize=(8,5))
plt.bar(age_labels, age_counts, color='skyblue', edgecolor='black')
plt.title("Customer Age Group Distribution")
plt.xlabel("Age Groups")
plt.ylabel("Number of Customers")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
months = np.arange(1,13)

month_counts = [np.sum(purchase_months == m) for m in months]

plt.figure(figsize=(8,5))
plt.bar(months, month_counts, color='orange', edgecolor='black')
plt.title('Monthly Purchase Trend')
plt.xlabel('Month')
plt.ylabel('Number of Purchases')
plt.xticks(months)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
labels = ['One-time Customers', 'Repeat Customers']
sizes = [num_customers - total_repeat, total_repeat]
colors = ['lightcoral', 'lightgreen']

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, explode=(0.05,0))
plt.title('Customer Loyalty Distribution')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(amount_spent, bins=30, color='purple', edgecolor='black')
plt.title('Spending Distribution per Customer')
plt.xlabel('Amount Spent ()')
plt.ylabel('Number of Customers')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
